### Legends in matplot

```python
# adding legends to our data visualization

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,12, 12)
```

```python
fig = plt.figure() # ---> creats Figur obj canvas

axes = fig.add_axes([0,0,1,1])  # ---> adding in axes to figure object with left corner at (0,0) & full width & full height equals 1

axes.plot(x,x)  # ----> x vs x plot
axes.plot(x, x**2)  # ---> x vs x**2 plot

#NB --> tthis plots will be hard to backtrack & tell one plot is x vs x & other is x vs x**2

#----> i.e, we can add in *legends to tell what each color represents 
```

```python
# Adding Legends in our axes

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])  # ---> adding in axes to figure object with left corner at (0,0) & full width & full height equals 1

axes.plot(x,x, label= 'X vs X')  # ----> x vs x plot ---> passing in label String  we want our legends to associate with this plots
axes.plot(x, x**2, label='X vs X**2')  # ---> x vs x**2 plot ---> pass in label string to asssociate with legends

#axes.legend()  # ----> adding legends inside our axes by default (which can then overlap with plots in axes)

#axes.legend(loc='upper rigth') ---> legends on top right inside axes ----> * refer doc - shift + Tab

axes.legend(loc=(-.3, 0.8)); # ---> placing legends manually outside the axes passing left corner tuple for (x,y) co-ordinate  
```

```python
# ---> ** Recommend to add in labels legends if we have multiple plotting in same axes

```

### Visual Styling - Colors & styles

```python
# Visual styles ----> line style, line widht, markers, colors etc 

import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,12,12)
x
```

```python
#-----> we learnt to add title, label, axis limit on axes (Figur object)

#------> we ll learnt to change defaukt line color & style
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color = "blue") # ---> plots X vs X  with blue color line
```

```python
#passing in custom color ----> we can pass in RGB hexcode

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
 
axes.plot(x, x, color = '#fcde70', label= 'X vs X')  # passing in custom hex code for rgb 
axes.plot(x, x+3, color= 'blue', label='X vs X+3')
axes.legend(loc=(1.1, 0.5))
```

```python
# Line & marker style

# line width ---> customize line width

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# line style optionos -->  --, -, -., :


axes.plot(x, x, color = '#fcde70', label= 'X vs X', lw =5, ls = '--') # passing in line wideth 5 times the default line width & dash line style
axes.plot(x, x+3, color= 'blue', label='X vs X+3', linewidth = 0.5, ls='-.') # passing in line width half of the default width & dash dot line style
axes.plot(x, x**2, color= "orange", label="X vs X*2", lw=2, ls= ':')                                                           

axes.legend(loc=(1.1, 0.8))

#NB ---> we can do linewidth or lw for short to set width of line & same for ls & linestyle
```

```python
# custom line style ----> we can do custom line stlye

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# line style optionos -->  --, -, -., :


lines = axes.plot(x, x, color = '#fcde70', label= 'X vs X', lw =5, ls = '--') # passing in line wideth 5 times the default line width & dash l
line_two = axes.plot(x, x**2, color= "orange", label="X vs X*2", lw=2, ls= ':')


lines[0].set_dashes([5,2,10,2])  # ---> we index line object from lines list, call .set_dashe() & pass in list that correspond with our custom das sequence

line_two[0].set_dashes([1,5,10,1,5,1])

#NB - we pass how many point we want, then how many pinst we space after it --> eg - 5 solid pts then 2 space pts and then 10 solid pts & then 2 blank pts

#NB ---> i.e custom dash line pts comes in pair (solid & space and so on...)
```

```python

# plotting omly poinst that occur
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])


# line style optionos -->  --, -, -., :


axes.plot(x, x, color = '#fcde70', label= 'X vs X', lw =2, ls= '--', marker= 'o', ms=10)  # passing in marker "o" & size #---> we can see pt marker 0f each x pt
axes.plot(x, x**2, color = 'red', label= 'X vs X', lw =1, ls= '--', marker= '+', markersize = 15) # passing in markersize 15 times the size 
```

```python
# center & edge ---> we can edit its edge width, color & face color

fig = plt.figure()

axes = fig.add_axes([0,0,1,1])


# line style optionos -->  --, -, -., :


axes.plot(x, x, color = '#fcde70', label= 'X vs X', lw =2, ls= '--', 
          marker= 'o', ms=10, markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange') 
 #----> passing in marker face color, edge width, edge color (higgh level customization) 

```
