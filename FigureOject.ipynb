# Implementing Figure Object

```python
#Implementing Figure & Axes

import numpy as np
import matplotlib.pyplot as plt
```

```python
#DATA --> using numpy

a = np.linspace(0,10, 11)
```

```python
a
```

```python
b = a ** 4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
x
```

```python
y = 2*x
```

```python
y
```

```python
fig = plt.figure()  #----> creating empty canvas 

axes = fig.add_axes([0,0,1,1])  # ----> creating canvas with full width & height of canvas (1,1) & bottom left at (0,0) #---->running this cell we can sse this axes 
#-->( detemining the location & sze of the axis)

# we can con now plot on this axes
axes.plot(x,y);  # ----> plot relation between x & y on this axes canvs

plt.show() #----> use this if working .py script to view the plor

#NB- all of this codes shoutld be run on same cell 
```

```python
# plot rel between a & b

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

```python
type(fig) # ---> figure object, i.e we can add many axes on it & plot on each axes

#NB --> by default/ convention, we call/nme this figure object as fig. *But -->for axes in fig we add label as axes1, axes2 etc (since there can be many axes in same fig)  
```

```python
# adding multiple axes to same Figure object (canvas)

fig = plt.figure()

# Large axes
axes = fig.add_axes([0,0,1,1])  #--> full widht & height as axes
axes.plot(a,b)  #------> plotting rel bet a & b

#Small axes 
axes = fig.add_axes([0.5,0.5, 0.25, 0.25 ]) #---> takes (0.2x, 0.2y) as left corner &  height & width are 50% of fig size 
axes.axes.plot(x,y)  # -----> plot rel bet x & y

#NB----> play around with this axes position --> we have (x,y) co-ord of bottom left corne & then (w,h) width & height of the axes 

plt.show()
```

```python
# zoom in section of the same plot as insert plot
# adding labels & titles to axes

fig = plt.figure()

#large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

#labels
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('relation between a & b')

#small axes
axes2 = fig.add_axes([0.5,0.5,0.25,0.2])
axes2.plot(a,b)

#labels
axes2.set_xlim(1,2)
axes2.set_ylim(0,500)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()


```

```python
#NB- i.e we can as many axes as we want & we can place them whwerever we want

#place zoomed plot outside the figure obj frame 

fig = plt.figure()

#large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

#labels
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('relation between a & b')

#small axes
axes2 = fig.add_axes([1,1,0.25,0.25])  # ---> left corner co-ord at (1,1) --> i.e at top right corner & hence out of the fig obj frame 
axes2.plot(a,b)

#labels
axes2.set_xlim(1,2)
axes2.set_ylim(0,500)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```

```python
# legends ---> plots outside of main Figure object plot

# we can also increse the size of axes as we may want

fig = plt.figure()

#large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

#labels
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('relation between a & b')

#small axes
axes2 = fig.add_axes([1,1,2,2])  # ---> increse height & widht, double the size of main figure object axes
axes2.plot(a,b)

#labels
axes2.set_xlim(1,2)
axes2.set_ylim(0,500)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```
