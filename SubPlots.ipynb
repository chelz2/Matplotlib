# Subplots functionality in MATPLOT

```python
# Subplots ---> we can create multiple axes with plots to be side by side in single Figure canvas manually (last lesson)

#---> *But we can use builtin method .subplot() to create a subplots for us 

# .subplot() ---> allows us to create us to create Figure & Axes objects in side by side formation

#-------> we pass in no. of rows & columns we want for set of axes

# -----> returns both our Figure canvas & axes objects which is numpy array which we can then index to grab & plot on top of that axes

# by convention We label as (fig, axes) ---> fig is entire canvas & axes theh numpy array holding each axes objects

```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATAS

a = np.linspace(0,11,10)
b = a ** 4    # expotential rel btween a & b
```

```python
x = np.arange(0,11)
y = 2 * x  # linear rel bet x & y
```

```python
fig , axes = plt.subplots()

axes.axes.plot(a,b)  # ---> we can do plot in default single axes directly for single plot

# ---> for single 1 plot , we can manually create the figure (last lesson)
```

```python
# mutple axes 

fig, axes = plt.subplots(nrows=1, ncols=2) # ----> create two two axes in align 
```

```python
# Nb - the moment we ask for more than ! row & 1 column ----> axes is no longer a matplot axes object

type(axes)  # ---> we can see axes is now a numpy array
```

```python
# mutiple plots using subplots() method ----> plot x & y relation and a & b relation in seperate axes in same figure object

fig, axes = plt.subplots(nrows=1, ncols=2) # ----> creates two axes side by side  

axes[0].plot(x, y) # ----> first axes at index 0 is plots the x & y rel

axes[1].plot(a,b) # -----> # second axes plots a & b rel 
```

```python

fig, axes = plt.subplots(nrows=2,ncols=2)  # ----> creates 4 axes


axes[0].plot(a,b) # --> returns  attribute error
axes[1].plot(x,y) # ---> returns attribute error as this numpy array of axes is now 2dimension
```

```python
type(axes) 
```

```python
fig , axes = plt.subplots(nrows=2,ncols=2)  # ----> generates 2d array with 4 axes

axes[0][0].plot(a,b)  # ----> plots a & b in 0 index axes of 0 index array

axes[1][0].plot(x,y)  #-----> plots x & y relation on 0 index axes of 1 index arry
```

```python
# 1D along columns/rows array - axes

fig, axes = plt.subplots(nrows=3, ncols=1)  # ----> 1D array with 1 column & 3 rows 

axes[0].plot(y,x)
axes[2].plot(a,b)
```

```python
axes.shape 
```

```python
type(axes)
```

```python
# looping throug axes in 1 d array

fig , axes = plt.subplots(nrows=3,ncols=1)  # ----> generates 1d array with 3 rows & 1 column


for ax in axes:     # axes in 1 d array
    ax.plot(a,b)

plt.tight_layout()  # ---> cretes a space between axes (makes sure no tickers are overlapping creating spce around them)
```

```python
# manually spacing

fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a,b)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=0.8, hspace=0.8)  # width height spoce in ratio to avarage axes height & widht size ---> refer docs for more spacing features\ 
```

```python
# edit parameters of subplots ---> adding labels & titles 
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8), dpi =150) # ---> adding figsize & dpi (same as last lesson)

axes[0][0].plot(a,b)          # first axes
axes[0][0].set_xlabel = "A"  #---> adding X lbael 
axes[0][0].set_ylabel = "B"  #---> add y label
axes[0][0].set_xlim(2,10)   #----> add x limit
axes[0][0].set_ylim(2,5000)  #---> add y limit
axes[0][0].set_title('Rel bet a & b')

axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)


fig.suptitle(" Adding Figure Title ", fontsize=20)  # ---> addig fig title
  
fig.subplots_adjust(wspace=0.8, hspace=0.8)  # width height spoce in ratio to avarage axes height & widht size ---> refer docs for more spacing features\ 

fig.savefig("subplots.png", bbox_inches='tight')   #-----> saves this subplots as .png file
```
