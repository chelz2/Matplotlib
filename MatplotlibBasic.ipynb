```python
# Matplotlib Basics --->

#------> most basic way to use matpllib is by calling builtin function plot calls, eg - plt.plot(x,y)
#------> simple to use but dont allow for high level of control  * Recommend for quick visualize of relations in data only

#NB- for .py python script, use plt.shoW() to visualize the plot

import matplotlib.pyplot as plt
```

```python
import numpy as np 
```

```python

x = np.arange(0,10)
```

```python
y = 2*x  # -----> linear relationship bet x & y
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)

#NB --> use plt.show() at end of code to see the plot if on .py script
```

```python
# edit matplotlib fuctional command ----> eg - we can add X labe & Y label 
# Other Figure parameters ---> customizing the plots

plt.plot(x,y)  #----->plots relation between x datas & y datas
plt.title('Retation between x & y')  # ---> adding title to plot
plt.xlabel('X axis')  # ----> adds 'X axis' name to horizantal line 
plt.ylabel('Y axis')  # ----> adds 'Y axis' to vertical line
plt.xlim(0,6)   # -----> limiting lenght of x axis from 0 to 6 only
plt.ylim(0, 15)  # ----> limiting lenght of y axis from 0 to 15 only
plt.savefig('firstmatplot.png'); # -----> save this as this as a png file # NB - we can save plot otr image format also like .jpg etc;

#NB - all this block of codes should be run on same cell
```
